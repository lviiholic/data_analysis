### 3.9a Analyse the frequency of the words then set a threshold for visualization

The above word cloud has many repeated words that do not fully represent the frequency of the word. Therefore we reset the threshold for the word and filter it.

In the following steps, I will use stop words and remove the most frequent word "drama" and remove ",". Then set a threshold of 10 to filter low frequency words.

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter

# Convert 'Genres' column to strings
moviesDF['Genres'] = moviesDF['Genres'].astype(str)

# Combine all genre values into a single string
genres_text = ' '.join(moviesDF['Genres'])

# Tokenize the text into individual words
tokens = word_tokenize(genres_text)

# Remove stopwords (common words with little meaning)
stop_words = set(stopwords.words('english'))
stop_words.add(',')

# Since one movies may have many genres and most of them are Drama
# So drama is the most frequent word, we can remove it
tokens = [word for word in tokens if word.lower() not in stop_words and word.lower() != 'drama']

# Count the frequency of each word
word_counts = Counter(tokens)

# Set a threshold to filter less frequent words
threshold = 10

# Filter out words based on the threshold
filtered_words = {word: count for word, count in word_counts.items() if count >= threshold}

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(filtered_words)

# Plot the word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Movie Genres Word Cloud (updated)')
plt.show()
```

```python
# Create a df from the word_counts dictionary
frequency_df = pd.DataFrame.from_dict(word_counts, orient='index', columns=['Count'])

# Sort the df by count in descending order
frequency_df = frequency_df.sort_values(by='Count', ascending=False)

frequency_df
```

The word cloud map is now clearer with the updated version, with each word appearing only once and no repetitions. The most frequent movie genres are Comedy: 1836, Thriller: 1339, Action: 1076, Documentary: 1024.

## RQ2 (Pivot table): Overview of the Movie Markets in the Top Five Countries

```python
# Group the df by country and count the number of movies
top_5_countries = moviesDF['Country'].value_counts().head(5).index.tolist()
filtered_movies = moviesDF[moviesDF['Country'].isin(top_5_countries)]

# Create the pivot table with average rating, average runtime, and movie counts
pivot_table = pd.pivot_table(filtered_movies,
                             values=['IMDb', 'Runtime','ID'],
                             index='Country',
                             aggfunc={'IMDb': 'mean', 'Runtime': 'mean', 'ID': 'count'})

pivot_table = pivot_table.rename(columns={'IMDb': 'Average Rating',
                                          'Runtime': 'Average Runtime',
                                          'ID': 'Movie Count'})
# Get the common language
languages = filtered_movies.groupby('Country')['Language'].apply(lambda x: x.value_counts().idxmax())
pivot_table['Common Language'] = languages

# Get the common genre
genres = filtered_movies.groupby('Country')['Genres'].apply(lambda x: x.str.split(',').explode().mode()[0])
pivot_table['Common Genre'] = genres

# Display
pivot_table
```

This pivot table compares the 5 countries with the highest number of movies and their different movie characteristics. As we can see, the five countries with the highest number of movies are **the United States, India, the United Kingdom, Canada and Australia.** Since a movie may be classified in more than one country, the U.S. has an estimated 3,501+ movies in the dataset, nearly half of the overall movies and far more than the second place. India has 430 movies ranking second, while the UK ranks third.

Of these five countries, the language of the movie is English except India, where it is Hindi. **This highlights the dominance of English-language movies in the global movie market and its international availability.** Meanwhile, the most common movie genre in all five countries is Drama, which shows that Drama is very popular in the global movie market.

Among of them, **Indian movies have the highest average IMDb rating and average runtime, and the average running time of each movie is more than 2 hours.**This may indicate the higher quality and uniqueness of Indian movies in the market. The average running time of movies in the other four countries is less than one and a half hours. Canada movies have the lowest average IMDb rating and average running time.

These details provide a more comprehensive understanding of the movie landscape and characteristics of movies from each of the top five countries. Users can select movies from different countries if they have a preference for a specific language, genre, movie length, or quality rating.
